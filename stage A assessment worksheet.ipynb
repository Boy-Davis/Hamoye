#### Question 1: Select columns â€˜Y2017â€™ and â€˜Areaâ€™, Perform a groupby operation on â€˜Areaâ€™. Which of these Areas had the 7th lowest sum in 2017?

```python
food_by_area = food_df.loc[:, ['Y2017', 'Area']]
```
